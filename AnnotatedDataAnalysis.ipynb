## Data loading
---
In order to perform data analysis on annotated dataset just download whole dataset form Google Drive, select only annotated rows and save it as .csv. Then perform infos and annotations merge by new indices.

```python
df_infos = pd.read_csv('data/sady_main/sady_infos_raw.csv')
df_infos.head(2)
```

```python
df_infos.columns
```

```python
df_annotated = pd.read_csv('data/sady_main/sady_date_annotated.csv', sep='\t')
df_annotated = df_annotated.drop(columns=['date', 'time'])
df_annotated.head(2)
```

Get only useful data from infos dataframe.

```python
df_infos = df_infos[['new_id', 'date', 'time']]  # and maybe other ... later
df_infos = df_infos.dropna()
df_infos['new_id'] = df_infos['new_id'].astype(int)
df_infos
```

```python
df_combined = df_infos.merge(df_annotated, left_on='new_id', right_on='id')
df_combined = df_combined.drop(columns=['new_id'])
df_combined
```

## Yearly counts analysis
---

```python
dates = df_combined['date']
dates.values
```

```python
years = np.array([d.split('-')[0] for d in dates.values])
years
```

```python
np.unique(years, return_counts=True)
```
